Import Libraries and Data Procesing

```python
import pandas as pd
```

```python
data=pd.read_csv("covtype.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum()
```

Make Feature and Labels

```python
X=data.drop("Cover_Type",axis=1)
Y=data["Cover_Type"]
```

Spliting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

Model

```python
from sklearn.ensemble import RandomForestClassifier
```

Model Train

```python
model=RandomForestClassifier(max_depth=17,random_state=42,n_estimators=300)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_Predict=model.predict(X_test)
```

Testing Model Score

```python
from sklearn.metrics import accuracy_score
```

```python
print("Train Score : ",model.score(X_train,Y_train))
print("Test Score : ",model.score(X_test,Y_test))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

importances = model.feature_importances_

feature_names = X.columns

feat_imp = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

feat_imp_sorted = feat_imp.sort_values(by='Importance', ascending=False)

print("Top 10 Important Features:")
print(feat_imp_sorted.head(10))
sns.barplot(x='Importance', y='Feature', data=feat_imp_sorted.head(10))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

cm = confusion_matrix(Y_test, Y_Predict)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(Y_test),
            yticklabels=np.unique(Y_test))

plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix for Forest Cover Type Classification")
plt.tight_layout()
plt.show()
```
